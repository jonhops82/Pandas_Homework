```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Calculate the number of unique players
player_count = len(purchase_data_df["SN"].unique())
print(f"There are a total of {player_count} unique players.")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate the number of unique items
unique_items = len(purchase_data_df["Item ID"].unique())
print(f"There are a total of {unique_items} unique items.")

# Calculate the average price of all items
avg_item_price = purchase_data_df["Price"].mean()
print(f"The average price of purchased items is ${round(avg_item_price,2)}.")

# Calculate the total number of purchases
total_purchases = len(purchase_data_df["Price"])
print(f"There are a total of {total_purchases} purchases.")

# Calculate the total revenue
total_revenue = purchase_data_df["Price"].sum()
print(f"The total revenue from all purchases is ${total_revenue }.")
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed



```python
# group by "Gender" column based on unique "SN" column
# calculates the number of males and females based on unique "SN" column
# basically a countif where we care counting "Gender" if "SN" is unique
gender_data_df = purchase_data_df.groupby(["Gender"])["SN"].nunique()

# pulling and naming variables from the table
male_users = gender_data_df.loc["Male"]
female_users = gender_data_df.loc["Female"]
other_users = gender_data_df.loc["Other / Non-Disclosed"]
total_users = male_users + female_users + other_users

# calculating percentages and naming as variables
male_percent = round(100* (male_users / total_users),2)
female_percent = round(100* (female_users / total_users),2)
other_percent = round(100* (other_users / total_users),2)
total_percent = 100.00

# create a dictionary of data to be presented
gender_dictionary = {"Gender": ["Male", "Female", "Non-Disclosed", "Total"],
                  "Count": [male_users, female_users, other_users, total_users],
                  "Percent": [male_percent, female_percent, other_percent, total_percent]}

# read dictionary into a dataframe to present
gender_table_df = pd.DataFrame(gender_dictionary)
gender_table_df
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data_df.groupby('Gender').Price.count()
purchase_average = purchase_data_df.groupby('Gender').Price.mean()
purchase_total = purchase_data_df.groupby('Gender').Price.sum()
 

prices_df = pd.DataFrame({"Count": purchase_count,
                         "Average Purchase Price": round(purchase_average,2),
                         "Total Purchase Value": purchase_total,
                         "Average Total Purchase per Person": round(
                             (purchase_total) / (purchase_data_df.groupby(["Gender"])["SN"].nunique())
                                                ,2)})
prices_df


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data_df['age_bins'] = pd.cut(x=purchase_data_df['Age'],bins=[0,9.99,14,19,24,29,34,39,120])

age_data_df = purchase_data_df.groupby(["age_bins"])["SN"].nunique()

ages_dictionary = {"Age Range": ["0-10", "10-14", "14-19", "19-24", "24-29", "29-34", "34-39", "39+"], 
                   "Total Count": age_data_df,
                   "Percent of Players": round(100* (age_data_df / total_users),2)}

ages_df = pd.DataFrame(ages_dictionary)
ages_df



```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_count = purchase_data_df.groupby(["SN"])['Price'].count()
SN_mean = purchase_data_df.groupby(["SN"])['Price'].mean()
SN_sum = purchase_data_df.groupby(["SN"])['Price'].sum()


ages_df2 = pd.DataFrame({"Purchase Count": SN_count,
                         "Average Purchase Price": round(SN_mean,2),
                         "Total Purchase Value": SN_sum})

ages_sort_df2 = ages_df2.sort_values(by=["Total Purchase Value"], ascending=False)

ages_sort_df2.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_id = purchase_data_df.groupby(["Item ID", "Item Name"])['Item ID'].mean()
item_name = purchase_data_df.groupby(["Item ID", "Item Name"])['Item Name'].unique()
item_count = purchase_data_df.groupby(["Item ID", "Item Name"])['Price'].count()
item_mean = purchase_data_df.groupby(["Item ID", "Item Name"])['Price'].mean()
item_sum = purchase_data_df.groupby(["Item ID", "Item Name"])['Price'].sum()


item_df = pd.DataFrame({ "Purchase Count": item_count,
                         "Item Price": round(item_mean,2),
                         "Total Purchase Value": item_sum})

item_sort_df = item_df.sort_values(by=["Purchase Count"], ascending=False)

item_sort_df.head(5)
```
